```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
column_to_drop = data.iloc[:, 2].name
X1=X1.drop(columns=[column_to_drop])
Y=X1.iloc[:,2]
```

```python
Y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X1,Y,test_size=0.2)
```

```python
X_train=x_train.drop(columns=["Geography"])
X_test=x_test.drop(columns=["Geography"])
```

```python

from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit_transform(X_train)
scaler.fit_transform(X_test)
```

```python
# Initialize the LinearRegression model
model = LinearRegression()

# Fit the model on the training data
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)

```

```python
# Calculate Mean Squared Error (MSE)
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Calculate R-squared (R2) score
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```

```python
# Coefficients (slope) for each feature
coefficients = model.coef_
print(f'Coefficients: {coefficients}')

# Intercept (constant term)
intercept = model.intercept_
print(f'Intercept: {intercept}')
```

```python
# Evaluate the model
accuracy = model.score(X_test, y_test)
print(f"Model accuracy: {accuracy:.2f}")
```

```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation
cv_scores = cross_val_score(model, X_train, y_train, cv=5)
print(f"Cross-validated accuracy: {cv_scores.mean():.2f}")
```
