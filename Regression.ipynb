```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("C:/Users/vanim/Downloads/canada_per_capita_income.csv",index_col=False)
data.head()
```

```python
plt.scatter(data["year"],data["per capita income (US$)"],color="red",marker="*")
plt.xlabel("year" ,fontsize=20)
plt.ylabel("per capita income (US$)" ,fontsize=20)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X=data[["year"]]
Y=data.iloc[:,-1]
```

```python
model=LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.predict([[2020]])
```

```python
b=model.coef_
b
```

```python
m=model.intercept_
m
```

```python
model.score(X,Y)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
# Degree of the polynomial
degree = 5

# Create a pipeline that first transforms the features and then fits a linear regression model
model = Pipeline([
    ('poly_features', PolynomialFeatures(degree=degree)),
    ('linear_regression', LinearRegression())
])
```

```python
model.fit(X, Y)
```

```python
model.predict([[2020]])
```

```python
plt.scatter(data["year"],data["per capita income (US$)"],color="red",marker="*")
plt.xlabel("year" ,fontsize=20)
plt.ylabel("per capita income (US$)" ,fontsize=20)
plt.plot(X, model.predict(X), color="blue")
plt.show()
```

```python
model.score(X,Y)
```
